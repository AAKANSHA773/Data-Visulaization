```python
import seaborn as sns
from scipy.stats import norm
```

```python
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.distplot(tips ["size"])
```

```python
sns.distplot(tips["tip"])
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"],hist=False)#remove background
```

```python
sns.distplot(tips["total_bill"],kde=False)#remove line
```

```python
sns.distplot(tips["total_bill"],rug=True)
```

```python
sns.distplot(tips["total_bill"],fit=norm)#normalaize lineplot
```

```python
sns.distplot(tips["total_bill"],color="r")
```

```python
sns.distplot(tips["total_bill"],vertical=True)
```

```python
sns.distplot(tips["total_bill"],norm_hist=True)
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(tips["total_bill"],bins=bins)
plt.xticks(bins)
sns.set()
```

```python
plt.figure(figsize=(16,9))
sns.distplot(tips["total_bill"],
             hist_kws={'color':'#DC134C',"edgecolor":"#aaff00",
                       'linewidth':5,'alpha':0.9})
```

```python
plt.figure(figsize=(16,9))
sns.distplot(tips["total_bill"],
             hist_kws={'color':'#DC134C',"edgecolor":"#aaff00",
                       'linewidth':"5",'alpha':0.9},
             kde_kws={'color':'#8e00ce',
                      'linewidth':8,'alpha':0.9},
            rug=True,
            rug_kws={'color':'#8e00ce',
                      'linewidth':8,'alpha':0.9} )
```
