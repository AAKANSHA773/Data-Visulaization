```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=sns.load_dataset("titanic")
titanic
```

```python
sns.scatterplot(x="age",y="fare",data=titanic)
```

```python
sns.scatterplot(x="age",y="fare",data=titanic,hue="sex")
```

```python
sns.scatterplot(x="age",y="fare",data=titanic,hue="sex",style="who")
```

```python
sns.scatterplot(x="age",y="fare",data=titanic,hue="sex",
                style="who",size="who",sizes=(100,400))
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x="age",y="fare",data=titanic,hue="sex",
                style="who",size="who",sizes=(100,400))
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x="who",y="fare",data=titanic,hue="alive",
                style="alive",size="who",sizes=(100,400))
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x="who",y="fare",data=titanic,hue="alive",
                style="alive",size="who",sizes=(100,400),
               palette="inferno_r",alpha=.7)
```
