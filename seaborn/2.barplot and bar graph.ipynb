```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.barplot(x=tips["day"],y=tips["total_bill"])
```

```python
sns.barplot(x=tips["day"],y=tips["total_bill"],hue=tips["sex"])
```

```python

sns.barplot(x="day",y="total_bill",hue="sex",data=tips)
#data use to we are not use datafrane agian and again
```

```python
order = ["sun", "thue", "fri", "sat"]

sns.barplot(x="day",y="total_bill",hue="sex",
            data = tips, order = order)
# i don't no why are data not show
```

```python
sns.barplot(x="day",y="total_bill",hue="sex",data=tips,
            estimator=np.max)
```

```python
sns.barplot(x="day",y="total_bill",hue="sex",data=tips,
           estimator=np.max,ci=31)
```

```python
sns.barplot(x="day",y="total_bill",hue='sex',data=tips,palette="hot")
```

```python
sns.set()
kwargs={'alpha':0.9,'linestyle':':','linewidth':5,'edgecolor':'black'}
sns.barplot(x="day",y="total_bill",data=tips,**kwargs)
```

```python
#two type use to keyword argument
sns.barplot(x="day",y="total_bill",data=tips,alpha=0.9,
            linestyle="-.",linewidth=6,edgecolor="g")
```

```python
ax=sns.barplot(x="day",y="total_bill",data=tips,alpha=0.9,
            linestyle="-.",linewidth=6,edgecolor="g")
ax.set(title="barplot of topic DataFrame",
      xlabel="Days",
      ylabel="Total bill")
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x="day",y="total_bill",data=tips,alpha=0.9,
            linestyle="-.",linewidth=6,edgecolor="g")
plt.title("barplot day and totle bill",fontsize=20)
plt.xlabel("Days",fontsize=20)
plt.ylabel("totle bill",fontsize=20)
plt.show()
```
