```python
data=pd.read_csv("Who_is_responsible_for_global_warming.csv")
```

```python
data
```

```python
new_data=data.drop(["Country Code","Indicator Name",
                    "Indicator Code"],axis=1).set_index("Country Name")
new_data.head()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(new_data)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(new_data,vmin=0, vmax=21)
```

```python
plt.figure(figsize=(13,7))
sns.heatmap(new_data,vmin=0, vmax=21,cmap="coolwarm",annot=True)
```

```python
annot_arr=np.array([['a00','a01','a02'],['a03','a04','a05'],['a06','a07','a08'],['a09','a10','a11'],['a013','a14','a15']])
```

```python
annot_arr
```

```python
sns.heatmap(arr,annot=annot_arr,fmt="s")
```

```python
plt.figure(figsize=(16,9))
annot_kws={'fontsize':8, 
          "fontstyle":"italic",
         "color":"black",
         "alpha":0.9,
         "rotation":"vertical",
         'verticalalignment':"center",
         "backgroundcolor":"w"}
sns.heatmap(new_data,vmin=0, vmax=24,cmap="coolwarm",annot=True,annot_kws=annot_kws)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(new_data,vmin=0, vmax=21,cmap="coolwarm",annot=True,linewidth=4)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(new_data,vmin=0, vmax=21,cmap="coolwarm",annot=True,linewidth=4,linecolor="black")
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(new_data,cbar=False,xticklabels=False,yticklabels=False)
```

```python
plt.figure(figsize=(16,9))
cbar_kws={"orientation":"horizontal",
         'shrink':1,
         'extend':'min',
          'extendfrac':0.1,
         'ticks':np.arange(0.22),
         'drawedges':True}
sns.heatmap(new_data,cbar_kws=cbar_kws)
```

```python
plt.figure(figsize=(16,9))
contry_code=['USA','GBR','IND','CHN','RUS','AUS','FRA','DEU','CAN','BRA','ARG','P','NP','BAN','JP']
sns.heatmap(new_data,xticklabels=np.arange(0,15),yticklabels=contry_code)
```

```python
plt.figure(figsize=(16,9))
AX=sns.heatmap(new_data)
AX.set(title="heatmap of who is responsible for globle warming",
       xlabel="CO@emissio per year",
      ylabel="contry name")
sns.set(font_scale=5)
```
