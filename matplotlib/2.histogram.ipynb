```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age=np.random.randint(18,45,(100))
py_student_age=np.random.randint(15,40,(100))
```

```python
ml_student_age
py_student_age# it mix call age. so we use print function
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.title("ml_student_age_histo")
plt.xlabel("student age category")
plt.ylabel("number of student age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,color="black",
         label="ml_student")#rwidth give space to the bar
plt.title("ml_student_age_histo")
plt.xlabel("student age category")
plt.ylabel("number of student age")
plt.legend()
plt.show()
```

```python
   bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,color="black",
         label="ml_student")

plt.hist(py_student_age,bins,rwidth=0.8,color="r",
         label="py_student")

plt.title("ml_student_age_histo")
plt.xlabel("student age category")
plt.ylabel("number of student age")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,9))#figure size
bins=[15,20,25,30,35,40,45]
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,
         color=["black",'r'],
         label=["ml_student","py student"])

plt.title("ml_student_age_histo")
plt.xlabel("student age category")
plt.ylabel("number of student age")
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
bins=[15,20,25,30,35,40,45]
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,
         color=["black",'r'],
         label=["ml_student","py student"])

plt.title("ml_student_age_histo")
plt.xlabel("student age category")
plt.ylabel("number of student age")
plt.legend()
plt.show()
```
