# bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=['python','R','Al','Ml','DS']
class1_student=[30,20,10,25,10]#out of 100
class2_student=[40,5,30,30,10]
class3_student=[35,5,30,15,15]
```

```python
plt.bar(classes,class1_student)
```

```python
plt.barh(classes,class1_student)
```

```python
plt.bar(classes,class1_student,width=0.2,align="edge",color='r',
        edgecolor='black',linewidth=5,alpha=0.7,linestyle="--",
        label=" class1")
```

```python
plt.bar(classes,class1_student,visible=False)#hide the graph
```

```python
plt.figure(figsize=(16,9))
style.use('ggplot')
plt.bar(classes,class1_student,width=0.6,color="black",
        edgecolor="r",linewidth=3,alpha=0.9,linestyle=":")
```

```python
plt.barh(classes,class1_student,color="black",
        edgecolor="r",linewidth=3,alpha=0.9,linestyle=":")
```

```python
plt.bar(classes,class1_student,width=0.6,color="black",
        edgecolor="r",linewidth=3,alpha=0.9,linestyle=":")
plt.title('bar chart of iaip class',fontsize=20)
plt.xlabel("classes",fontsize=20)
plt.ylabel('No. of student',fontsize=20)
```

```python
plt.bar(classes,class1_student,width=0.6,color="black",label='black')

plt.bar(classes,class2_student,width=0.6,color="r",label='r')

plt.bar(classes,class3_student,width=0.6,color="g",label="g")

plt.title('bar chart of iaip class',fontsize=20)
plt.xlabel("classes",fontsize=20)
plt.ylabel('No. of student',fontsize=20)
plt.legend()
```

```python
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_student,width,color="black",label='class1 student')

plt.bar(classes_index+width,class2_student,width,color="r",label='class2 student')

plt.title('bar chart of iaip class',fontsize=20)
plt.xlabel("classes",fontsize=20)
plt.ylabel('No. of student',fontsize=20)
plt.legend()
```

```python
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_student,width,color="black",label='class1 student')

plt.bar(classes_index+width,class2_student,width,color="r",label='class2 student')

plt.bar(classes_index+width+width,class3_student,width,color="g",label='class3 student')

plt.xticks(classes_index+width,classes,rotation=20)#show name of class
plt.title('bar chart of iaip class',fontsize=20)
plt.xlabel("classes",fontsize=20)
plt.ylabel('No. of student',fontsize=20)
plt.legend()
```
