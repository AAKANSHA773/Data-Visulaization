```python
play_stor= pd.read_csv("googleplaystore.csv",nrows=1000)
play_stor.shape
```

```python
x=play_stor["Rating"]
y=play_stor["Installs"]
```

```python
plt.scatter(x,y)
plt.title("Google play stor app")
plt.xlabel("rating")
plt.ylabel("Install")
```

```python
plt.figure(figsize=(12,6))
plt.scatter(x,y, c="r", marker="*")
plt.title("Google play stor app")
plt.xlabel("rating")
plt.ylabel("installs")
```

```python
plt.figure(figsize=(12,6))
plt.scatter(x,y, c="r",marker="*",s=100,linewidths=10)
plt.title("Google play stor app")
plt.xlabel("rating")
plt.ylabel("installs")
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c="r", marker="*", s=100, linewidths=10)

plt.scatter(x,play_stor["Reviews"], c="y", marker="o",
            s=100,  linewidths=10)
plt.title("Google play stor app")
plt.xlabel("rating")
plt.ylabel("Reviews,installs")
```
